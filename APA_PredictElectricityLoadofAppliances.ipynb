<div style="text-align: justify">The purpose of the Feature Selection process, as an essential process of any application that uses data mining algorithms, is to search for a subset of features by excluding features with little or no information for the prediction problem. [31] In this way, the performance of the model is boosted while, at the same time, the processing penalties are lowered. Feature Selection techniques are divided into filter and wrapper methods. The wrapper-based techniques are model-based, meaning that features are selected based on the output of a data mining algorithm, thus making it a much slower approach than the filter-based techniques [32]. That is because filter-based techniques are model-free and they look only at the calculated feature relevance scores [32]. In our work, we have selected our features by implementing the wrapper-based step forward feature selection method. As an input, we have to specify the number of features that we would like to select (n), the algorithm model, and the selection criteria. In our case, we selected n to be 13 for the Smart* dataset and 12 for the REFIT dataset, but it can be manually modified by the user. In the beginning, there are no features selected.
Each feature is then evaluated and only n features that lead to the best performing model are selected. In order to determine that, we use the defined evaluation criteria. Since this is a regression problem and we want to use the mean squared error metric, we set the scoring to ‘neg_mean_squared_error’.

```python
# Function for turning hours into timing of the days.
def hours2timing(x):
  if x in [22,23,0,1,2,3]:
    timing = 'Night'
  elif x in range(4, 13):
    timing = 'Morning'
  elif x in range(13, 18):
    timing = 'Afternoon'
  elif x in range(18, 22):
    timing = 'Evening'
  else:
    timing = 'X'
  return timing
# Function for turning months into seasons
def seasons(x):
  if x in [12,1,2]:
    season = 'winter'
  elif x in range(3,6):
    season = 'spring'
  elif x in range(6,9):
    season = 'summer'
  elif x in range(9,12):
    season = 'autumn'
  else:
    season = 'X'
  return season
```

```python
# Feature engineering as a function.
def feature_engineering(data):
  data['Month'] = data.index.month
  data['Weekday'] = data.index.weekday
  data['Hour'] = data.index.hour
  data['Week'] = data.index.week
  data['WeekofYear'] = data.index.weekofyear
  data["Year"]=data.index.year
  data["Is_Weekend"]= data["Weekday"].isin([5,6])
  data['Timing'] = data['Hour'].apply(hours2timing)
  data['Season'] = data['Month'].apply(seasons)
  data['Weekday_sin'] = np.sin(data['Weekday'] * (2 * np.pi / 7))
  data['Weekday_cos'] = np.cos(data['Weekday'] * (2 * np.pi / 7))
  data['Month_sin'] = np.sin(2 * np.pi * data['Month']/12)
  data['Month_cos'] = np.cos(2 * np.pi * data['Month']/12)
  data['Hour_sin'] = np.sin(2 * np.pi * data['Hour']/23)
  data['Hour_cos'] = np.cos(2 * np.pi * data['Hour']/23)
  data["Is_Weekend"] = data["Is_Weekend"].astype(int)  
  return data
```

```python
# Aggregating the data on the hour level. Using backward fill for missing values.
def aggregator(data):
  data_hour = data.resample('1H').mean().bfill()
  return data_hour
```

```python
# Forward feature selection for the data. Chosen feature names are returned as an output
def feature_selection(features,labels,number_of_features):
  sfs = SFS(RandomForestRegressor(),
        k_features=number_of_features,
        forward=True,
        floating=False,
        scoring = 'neg_mean_squared_error',
        cv = 0)

  sfs.fit(features, labels)
  feature_names = np.asarray(sfs.k_feature_names_)
  return feature_names
```

```python
# Train test split. Used in the load agent for ML algorithms. Defined seperately to incorporate chosen features while dividing data.
def Train_Test_Split(data,features,appliance):
   test_size=.2
   test_index = int(len(data.dropna())*(1-test_size))
   X_train, X_test = data[features].iloc[:test_index,], data[features].iloc[test_index:,]
   Y_train = data[appliance].iloc[:test_index,]
   Y_test =  data[appliance].iloc[test_index:,]
   return {'X_train': X_train, 'X_test': X_test, 'Y_train': Y_train,"Y_test": Y_test}
```

```python
# Standardization of the train and test data.
def data_scaler(X_train,X_test):
  scaler = StandardScaler()
  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)
  return {"X_train": X_train, "X_test": X_test}
```

```python
# Function for evaluating mean squared errors
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    mse = mean_squared_error(test_labels,predictions)
    return mse
```

```python
# Function for returning the MAE,MSE,RMSE and MAPE for the best models that are chosen. 
# It also creates the 24 hour load graph for the chosen best model. The first 24 observations in the test data is used for that. 
def evaluate_best(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    mae = mean_absolute_error(test_labels,predictions)
    mse = mean_squared_error(test_labels,predictions)
    rmse = mean_squared_error(test_labels, predictions, squared=False)
    print("The best model is:"+"\n")
    print(model,'\n')
    print('Mean Absolute Error             : {:0.4f}'.format(mae))
    print('Mean Squared Error              : {:0.2f}'.format(mse))
    print('Root Mean Square Error          : {:0.2f}'.format(rmse))
    print('Mean Absolute Percentage Error  : {:0.2f}%'.format(mape))
    fig = plt.figure(figsize=(20,8))
    plt.plot(test_labels[0:24],label='Target Value',color='b')
    plt.plot(predictions[0:24],label='DL Model Prediction ', linestyle='--', color='y')
    plt.legend(loc=1)
```

```python
# A seperate function for returning the same outputs for machine learning models. 
def evaluate_best_ml(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    mae = mean_absolute_error(test_labels,predictions)
    mse = mean_squared_error(test_labels,predictions)
    rmse = mean_squared_error(test_labels, predictions, squared=False)
    print("The best model is:"+"\n")
    print(model,'\n')
    print('Mean Absolute Error             : {:0.4f}'.format(mae))
    print('Mean Squared Error              : {:0.2f}'.format(mse))
    print('Root Mean Square Error          : {:0.2f}'.format(rmse))
    print('Mean Absolute Percentage Error  : {:0.2f}%'.format(mape))
    fig = plt.figure(figsize=(20,8))
    plt.plot(test_labels.values[0:24],label='Target Value',color='b')
    plt.plot(predictions[0:24],label='ML Model Prediction ', linestyle='--', color='y')
    plt.legend(loc=1)
```

```python
# Function for determining the best model 
def most_accurate_model(models,test_features,test_labels):
  best=models[0]
  for model in models:
    if evaluate(model,test_features,test_labels)<evaluate(best,test_features,test_labels):
      best=model
  return best
```

```python
# Time series nested cross validation structure.
def time_series_nested_cv(data,appliance,number_of_features):
  n_splits = 5 # Number of train/cv/test folds
  trainTestSplit = TimeSeriesSplit(n_splits+1).split(data)
  next(trainTestSplit) # Skip the first fold
  y_index= data.columns.get_loc(appliance)# Getting the index of the target variable

  outer_results_rf = list() # Creating a list for the outer_results of random forest
  outer_results_gb = list() # Creating a list for the outer_results of gradient boosting
  outer_results_svr = list() # Creating a list for the outer_results of support vector regressor

  # Dividing the data to features and target variable
  X=data.drop(appliance, inplace=False, axis=1) 
  y=data[appliance]

  # Getting the feature names and creating the new feature space accordingly.
  feature_names = feature_selection(X,y,number_of_features)
  X=X[feature_names]

  
  for trainCvIndices, testIndices in trainTestSplit:
    # First, we split Train + CV and Test
    XTrainCv, yTrainCv = X.iloc[trainCvIndices,], data.iloc[trainCvIndices,y_index]
    XTest, yTest       = X.iloc[testIndices,]   , data.iloc[testIndices,y_index]

    # Then, we build a list of the form [ ( [...Train Indices...], [...CV Indices...]  )]
    testLength = len(XTest)
    trainCvSplit = [(list(range(trainCvIndices[0],trainCvIndices[-testLength])),
                     list(range(trainCvIndices[-testLength],trainCvIndices[-1]+1)))]
    
    # Scaling X and y
    scaled=data_scaler(XTrainCv,XTest)
    XTrainCv=scaled["X_train"]
    XTest=scaled["X_test"]

    # Define the first model: Random Forest Regressor
    rf_model = RandomForestRegressor(random_state=123)
    # Define the grid space for the model
    rf_param_grid = {'n_estimators': [100,200],
                      'max_features': [4, 6]}
    
    # Execute the grid search
    gs_rf = GridSearchCV(rf_model, rf_param_grid, cv=trainCvSplit, scoring='neg_mean_squared_error', verbose=0,n_jobs=-1)

    # Fit the grid search to predict
    rf_result = gs_rf.fit(XTrainCv, yTrainCv)
    
    # Get the best performing model with hyperparameters
    rf_best = rf_result.best_estimator_

    # Evaluate the model and store the mean squared error
    mse_rf=evaluate(rf_best,XTest,yTest)

    # Append the mean squared errors to the list
    outer_results_rf.append(mse_rf)
  	
  # Report progress - Kept as a comment to have a clean output can be also run to see the progress.
  # print('RF Holdout mse best params=%.3f,RF Best Cv mse=%.3f,RF Best params=%s' % (mse_rf, rf_result.best_score_, rf_result.best_params_))

    # Define the second model: Gradient Boosting Regressor
    gb_model = GradientBoostingRegressor(random_state=123)
   
    # Define the grid space for the model
    gb_param_grid = { 'n_estimators': [100,200,300],
                      'max_depth': [5,10],
                      'learning_rate': [0.01],}
    
    # Execute the grid search
    gs_gb = GridSearchCV(gb_model, gb_param_grid, cv=trainCvSplit, scoring='neg_mean_squared_error', verbose=0,n_jobs=-1)

    # Fit the grid search to predict
    gb_result = gs_gb.fit(XTrainCv, yTrainCv)
    
    # Get the best performing model with hyperparameters
    gb_best = gb_result.best_estimator_

    # Evaluate the model and store the mean squared error
    mse_gb=evaluate(gb_best,XTest,yTest)

    # Append the mean squared errors to the list
    outer_results_gb.append(mse_gb)

  	# Report progress - Kept as a comment to have a clean output can be also run to see the progress.
    #print('GB Holdout mse best params=%.3f,GB Best Cv mse=%.3f,GB Best params=%s' % (mse_gb, gb_result.best_score_, gb_result.best_params_))
   
    # Define the second model: Support Vector Regressor
    svr_model = svm.SVR()

    # Define the grid space for the model
    svr_param_grid = {'C': [10, 100],
                     'gamma': [1,0.1,0.01],
                     'kernel': ['rbf', 'sigmoid']}
    
    # Execute the grid search
    gs_svr = GridSearchCV(svr_model, svr_param_grid, cv=trainCvSplit, scoring='neg_mean_squared_error', verbose=0,n_jobs=-1)

    # Fit the grid search to predict
    svr_result = gs_svr.fit(XTrainCv, yTrainCv)
    
    # Get the best performing model with hyperparameters
    svr_best = svr_result.best_estimator_

    # Evaluate the model and store the mean squared error
    mse_svr=evaluate(svr_best,XTest,yTest)

    # Append the mean squared errors to the list
    outer_results_svr.append(mse_svr)

  	# Report progress - Kept as a comment to have a clean output can be also run to see the progress.
    # print('SVR Holdout mse best params=%.3f,SVR Best Cv mse=%.3f,SVR Best params=%s' % (mse_svr, svr_result.best_score_, svr_result.best_params_))

  # Calculating the mean of the MSE's for the 3 models
  result_rf=mean(outer_results_rf)
  result_gb=mean(outer_results_gb)
  result_svr=mean(outer_results_svr)

  #Storing the results in an array for comparison
  results=[result_rf,result_gb,result_svr]
  best=results[0]
  best_model=[]
  
  #Finding the best result in a for loop
  for result in results:
    if result < best:
      best = result
  
  # Assigning the best model according to the result of the for loop.
  if best==result_rf:
    best_model=rf_best
  elif best==result_gb:
    best_model=gb_best
  else:
    best_model=svr_best
  

  # Summarize the estimated performance of the 3 machine learning models
  print('Random Forest MSE: %.3f (%.3f)' % (mean(outer_results_rf), std(outer_results_rf)))
  print('Gradient Boosting MSE: %.3f (%.3f)' % (mean(outer_results_gb), std(outer_results_gb)))
  print('Support Vector Regressor MSE: %.3f (%.3f)' % (mean(outer_results_svr), std(outer_results_svr)))

    # Printing date ranges - kept as comment can be run in the first for loop to see the dates and the structure of the time series nested cross validation.
#    print("Training:"           , XTrainCv.index[0].date(), "--", XTrainCv.index[-testLength-1].date(),
#          ", Cv:"     , XTrainCv.index[-testLength].date(), "--", XTrainCv.index[-1].date(),
#          ", Test:"                , XTest.index[0].date(), "--", XTest.index[-1].date())

  return best_model
```

```python
# Splitting the data for the deep learning algorithms
def deeplearning_datasplit(prepped_data,appliance,isLSTM):
  
  # Shift target variable to first position
  first_column = prepped_data.pop(appliance)

  # Insert column using insert(position,column_name, first_column) function
  prepped_data.insert(0, appliance, first_column)
  df = prepped_data.apply (pd.to_numeric, errors='coerce')

  # Splitting the data to train and test set.
  x = df.iloc[:,1:].values
  y = df.iloc[:,0].values
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)

  # Standardization of the data
  sc = StandardScaler()
  x_train = sc.fit_transform(x_train)
  x_test = sc.transform(x_test)
  
  # The number of columns is extracted to use in reshaping
  shape_param=len(prepped_data.columns)-1
  
  # If the algorithm is LSTM then reshape the test data
  if(isLSTM):
      x_test = x_test.reshape((x_test.shape[0], 1, shape_param))
  return {'x_train': x_train, 'x_test': x_test, 'y_train': y_train,"y_test": y_test}
```

```python
# Definition of the ANN model
def ANN(x_train,y_train):
  ann = tf.keras.models.Sequential()
  ann.add(tf.keras.layers.Dense(units = 30, activation = 'sigmoid', kernel_regularizer=tf.keras.regularizers.l1(0.01)))
  ann.add(tf.keras.layers.Dense(units = 30, activation = 'sigmoid', kernel_regularizer=tf.keras.regularizers.l1(0.01)))
  ann.add(tf.keras.layers.Dense(units = 30, activation = 'sigmoid', kernel_regularizer=tf.keras.regularizers.l1(0.01)))
  ann.add(tf.keras.layers.Dense(units = 1))
  ann.compile(optimizer = 'adam',loss = 'mean_squared_error')
  ann.fit(x_train,y_train,batch_size = 128,epochs = 300, verbose=0)
  return ann
```

```python
# Definition of the LSTM model
def LSTM_MODEL(x_train,y_train):
  n_features=x_train.shape[1]
  x_train = x_train.reshape((x_train.shape[0],1, n_features))
  opt = Adam(lr=0.005)
  lstm = Sequential()
  lstm.add(LSTM(8, activation='relu',return_sequences=True, input_shape=(1, n_features),kernel_regularizer=tf.keras.regularizers.l1(0.01)))
  lstm.add(LSTM(8, activation='relu', input_shape=(1, n_features),kernel_regularizer=tf.keras.regularizers.l1(0.01)))
  lstm.add(Dense(1, activation='relu'))
  lstm.compile(loss='mean_squared_error', optimizer=opt)
  lstm.fit(x_train, y_train, epochs=100, batch_size=64, verbose=0)
  return lstm
```

```python
# The Load Agent

def load_agent(data,appliance,number_of_features):
  # DATA PREPERATION
  # The prepared data will be used for both machine learning algorithms and the deep learning algorithms.
  prepped_data=dummy_encoding(aggregator(feature_engineering(data)))

  # MACHINE LEARNING
  # Partitioning the data for feature selection
  X=prepped_data.drop(appliance, inplace=False, axis=1)
  y=prepped_data[appliance]
  # Feature selection for ML Algorithms
  feature_names = feature_selection(X,y,number_of_features)
  # Partitioning the data for ML Algorithm
  results=Train_Test_Split(prepped_data,feature_names,appliance)
  X_train=results["X_train"]
  X_test=results["X_test"]
  Y_train=results["Y_train"]
  Y_test=results["Y_test"]
  # Scaling the data for ML Algorithms
  scaled=data_scaler(X_train,X_test)
  X_train=scaled["X_train"]
  X_test=scaled["X_test"]
  # Applying time series nested cross validation and returning the best model with tuned hyperparameters
  best_ml_model=time_series_nested_cv(prepped_data,appliance,number_of_features)
  # Storing the MSE of best algorithm. Here, unlike the nested structure only MSE is calculated according to a 80-20% split.
  mse_best_ml=evaluate(best_ml_model,X_test,Y_test)
  
  #Evaluation of the best machine learning algorithm with all outputs. - Left as a comment can be executed to see the full results of best ML algorithm
  #evaluate_best_ml(best_ml_model,X_test,Y_test)

  #DEEP LEARNING
  # ANN Model
  results1=deeplearning_datasplit(prepped_data,appliance,False)
  x_train1=results1["x_train"]
  x_test1=results1["x_test"]
  y_train1=results1["y_train"]
  y_test1=results1["y_test"]
  model_ann=ANN(x_train1,y_train1)
  mse_ann=evaluate(model_ann,x_test1,y_test1)
  
  #Evaluation of the ANN algorithm with all outputs. - Left as a comment can be executed to see the full results of ANN algorithm 
  #evaluate_best(model_ann,x_test1,y_test1)

  # LSTM Model
  results2=deeplearning_datasplit(prepped_data,appliance,True)
  x_train2=results2["x_train"]
  x_test2=results2["x_test"]
  y_train2=results2["y_train"]
  y_test2=results2["y_test"]
  model_lstm=LSTM_MODEL(x_train2,y_train2)
  mse_lstm=evaluate(model_lstm,x_test2,y_test2)
  
  #Evaluation of the LSTM algorithm with all outputs. - Left as a comment can be executed to see the full results of LSTM algorithm
  #evaluate_best(model_lstm,x_test2,y_test2)
  
  # RESULT COMPARISON
  # Compare the results of the algorithms and return the best model
  # The results are put into an array for the loop
  results=[mse_best_ml,mse_ann,mse_lstm]
  best=results[0]
  best_model=[]
  
  # Lowest MSE is determined in the for loop.
  for result in results:
    if result < best:
      best = result
  
  # The best model is returned according to the output.
  if best==mse_best_ml:
    best_model=best_ml_model
    evaluate_best_ml(best_ml_model,X_test,Y_test)
  elif best==mse_ann:
    best_model=model_ann
    evaluate_best(model_ann,x_test1,y_test1)
  else:
    best_model=model_lstm
    evaluate_best(model_lstm,x_test2,y_test2)
  #return best_model
```
